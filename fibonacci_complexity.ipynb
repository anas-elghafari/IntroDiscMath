```python
def fib(n):
    if n== 0:
        return (0, 0)
    if n==1:
        return (1, 0)
    return add(fib(n-1), fib(n-2))

def add(t1, t2):
    return (t1[0] + t2[0], t1[1] + t2[1] + 1)
```

```python
%%time
for n in range(2, 39):
    f = fib(n)
    print("\nn now is: ", n)
    print("fib(", str(n), "): ", f[0])
    print("number of addition operations: ", f[1])
    print("ratio of #operations to fib: ", f[1]/f[0])
```

```python
%%time
for n in range(2, 39):
    f = fib(n)
    print("\nn now is: ", n)
    print("fib(", str(n), "): ", f[0])
    print("value of 1.618^n: ", 1.618**n)
    print("ratio of 1.618^n/fib(n) : ", (1.618**n)/f[0])
```
