```python
import math

def f(n):
    numlines = 0 #keeping track of number of lines printed
    for i in range(1, n+1):
        for j in range(1, n+1):
            print("i:" , i, "-- j:", j)
            numlines +=1
            j = math.floor(j/2)
            i = i+1
    print("NUMBER OF LINES PRINTED: ", numlines)
```

```python
f(5)
```

```python
f(10)
```
